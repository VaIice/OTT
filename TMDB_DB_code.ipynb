```python
######### 영화 콘텐츠 릴레이션 #########
import requests
import json
import pandas as pd

# 콘텐츠 엔티티
title_list = []
release_date_list = []
language_list = []
runtime_list = []
popularity_list = []
overview_list = []
id_list = []

# 콘텐츠-장르 엔티티
id_genre_list = []
genre_list = []

# 콘텐츠-플랫폼 엔티티
id_provider_list = []
provider_list = []

# 콘텐츠-제작국가 엔티티
id_country_list = []
country_list = []


# Test
start, end = 1, 10000

for i in range(start, end+1):
    try:
        # 영화 제공 플랫폼 get
        url = "https://api.themoviedb.org/3/movie/"+str(i)+"/watch/providers?api_key="
        response = requests.get(url)
        movie_provider_json = json.loads(response.text)
        movie_provider_result = movie_provider_json["results"]

        # 한국 지원 컨텐츠 필터링
        if "KR" in movie_provider_result:
            kr_movie_provider = movie_provider_result["KR"]

            # 영화 정보 get
            url = "https://api.themoviedb.org/3/movie/{}?api_key=language=ko-KR".format(
                i)
            response_sub = requests.get(url)
            movie = response_sub.text
            movie_json = json.loads(movie)

            # 영화 정보 출력
            if "title" in movie_json:
                Id = movie_json["id"]

                print("영화 아이디:",end=" ")
                print(Id)
                id_list.append(Id)

                print("영화 제목:", end=" ")
                print(movie_json["title"])
                title_list.append(movie_json["title"])

                print("영화 제작 국가:", end=" ")
                for j in movie_json["production_countries"]:
                    id_country_list.append(Id)
                    country_list.append(j['name'])
                print(country_list)

                print("영화 개봉일:", movie_json["release_date"])
                release_date_list.append(movie_json["release_date"])

                print("영화 언어:", movie_json["original_language"])
                language_list.append(movie_json["original_language"])

                print("영화 러닝타임:", movie_json["runtime"])
                runtime_list.append(movie_json["runtime"])

                print("영화 인기도:", movie_json["popularity"])
                popularity_list.append(movie_json["popularity"])

                print("영화 개요:", movie_json["overview"])
                overview_list.append(movie_json["overview"])

                print("영화 장르:", end=" ")
                for k in movie_json["genres"]:
                    id_genre_list.append(Id)
                    genre_list.append(k['id'])
                    print(k['id'], end=' ')
                print()

                # 구독, 대여, 구매 정보 확인

                print("영화 제공 : ", end="")
                providers = []
                if "flatrate" in kr_movie_provider:
                    flatrate = kr_movie_provider["flatrate"]
                    for provider in flatrate:
                        providers.append(provider["provider_id"])
                if "rent" in kr_movie_provider:
                    rent = kr_movie_provider["rent"]
                    for provider in rent:
                        providers.append(provider["provider_id"])
                if "buy" in kr_movie_provider:
                    buy = kr_movie_provider["buy"]
                    for provider in buy:
                        providers.append(provider["provider_id"])

                providers = set(providers)  # 중복 제거
                for p in providers:
                    id_provider_list.append(Id)
                    provider_list.append(p)
                print(providers)
                print()

    except:
        pass

df_movie = pd.DataFrame((zip(id_list, title_list, release_date_list, language_list, runtime_list, popularity_list, overview_list)), columns = ['콘텐츠번호', '콘텐츠제목', '콘텐츠개봉일', '콘텐츠언어', '콘텐츠길이', '콘텐츠인기도', '콘텐츠개요'])
df_movie_genre = pd.DataFrame((zip(id_genre_list,genre_list)), columns = ['콘텐츠번호', '장르번호'])
df_movie_country = pd.DataFrame((zip(id_country_list,country_list)), columns = ['콘텐츠번호', '국가번호'])
df_movie_provider = pd.DataFrame((zip(id_provider_list,provider_list)), columns = ['콘텐츠번호', '플랫폼번호'])
df_movie.to_csv(f'movie_{start}_{end}.csv', encoding='utf-8-sig', index=False)
df_movie_genre.to_csv(f'movie_genre_{start}_{end}.csv', encoding='utf-8-sig', index=False)
df_movie_country.to_csv(f'movie_country_{start}_{end}.csv', encoding='utf-8-sig', index=False)
df_movie_provider.to_csv(f'movie_provider_{start}_{end}.csv', encoding='utf-8-sig', index=False)
```

```python
import requests
import json
import pandas as pd

genre_ere = ['액션', '모험', '애니메이션', '코미디', '범죄', '다큐멘터리', '드라마', '가족', '판타지', '역사', '공포', '음악', '미스터리', '로맨스', 'SF', 'TV 영화', '스릴러', '전쟁', '서부']

movie_name_list = []
movie_index_list = []
genre_list = []
index = 0
genre_name_list = []

for i in range(0, 100):
    try:
        # 영화 제공 플랫폼 get
        url = "https://api.themoviedb.org/3/movie/"+str(i)+"/watch/providers?api_key="
        response = requests.get(url)
        movie_provider_json = json.loads(response.text)
        movie_provider_result = movie_provider_json["results"]

        # 한국 지원 컨텐츠 필터링
        if "KR" in movie_provider_result:
            # 영화 정보 get
            url = "https://api.themoviedb.org/3/movie/{}?api_key=&language=ko-KR".format(i)
            response_sub = requests.get(url)
            movie = response_sub.text
            movie_json = json.loads(movie)

            # 영화 정보 출력
            if "title" in movie_json:
              for genre in movie_json["genres"]:
                movie_name_list.append(movie_json["title"])
                genre_name_list.append(genre["name"])
                genre_list.append(genre_ere.index(genre["name"]))
                movie_index_list.append(index)

              index = index + 1

    except:
        pass

# 제목, 장르이름 삭제 예정
content_genre_df =  pd.DataFrame((zip(movie_name_list,genre_name_list, movie_index_list, genre_list)), columns = ['제목', '장르이름', '컨텐츠 번호', '장르 번호'])
content_genre_df
```
